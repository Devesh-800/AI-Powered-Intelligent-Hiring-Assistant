### Explanation

We use **pdfplumber** to extract text from PDF resumes.

```python
import pandas as pd
import numpy as np

import joblib

import pdfplumber

import re

from IPython.display import display

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

from config.py import API_KEY
```

```python
classifier = joblib.load("resume_classifier.pkl")

tfidf = joblib.load("tfidf.pkl")

encoder = joblib.load("label_encoder.pkl")
print("Models Loaded Successfully")
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
resume_file = list(uploaded.keys())[0]

resume_text = ""

with pdfplumber.open(resume_file) as pdf:

    for page in pdf.pages:

         text = page.extract_text()

         if text:
           resume_text += text
```

```python
print(resume_text[:1500])
```

```python
def clean_resume(text):

      text = re.sub(r"http\S+", " ", text)

      text = re.sub(r"www\S+", " ", text)

      text = re.sub(r"@\S+", " ", text)

      text = re.sub(r"[^A-Za-z ]", " ", text)

      text = re.sub(r"\s+", " ", text)

      return text.lower().strip()
```

```python
clean_text = clean_resume(resume_text)
```

```python
print(clean_text[:1000])
```

```python
resume_vector = tfidf.transform([clean_text])
```

```python
prediction = classifier.predict(resume_vector)

predicted_category = encoder.inverse_transform(prediction)

print("Predicted Category :")

print(predicted_category[0])
```

```python
print("Resume Successfully Classified.")
```

# Module 2 : Resume - Job Description Semantic Matching

## Objective

Traditional keyword matching is not enough because two resumes can use different words
while having the same meaning.

To solve this problem, we use **Sentence Transformers**.

Sentence Transformers convert complete sentences into dense numerical vectors (embeddings).

Then we calculate the **Cosine Similarity** between the Resume and Job Description.

Higher similarity means better match.

```python
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
print("Embedding Model Loaded Successfully")
```

```python
resume_embedding = embedding_model.encode(clean_text)
```

```python
resume_embedding.shape
```

```python
from google.colab import files

print("Upload Job Description")

uploaded_jd = files.upload()
```

```python
jd_file = list(uploaded_jd.keys())[0]
```

```python
if jd_file.endswith(".pdf"):

      jd_text = ""

      with pdfplumber.open(jd_file) as pdf:

           for page in pdf.pages:

                  text = page.extract_text()

                  if text:
                      jd_text += text

                  elif jd_file.endswith(".txt"):

                        with open(jd_file, "r", encoding="utf-8") as file:

                                jd_text = file.read()

                  else:

                        print("Unsupported File Format")
```

```python
print(jd_text[:1500])
```

```python
clean_jd = clean_resume(jd_text)
```

```python
jd_embedding = embedding_model.encode(clean_jd)
```

```python
jd_embedding.shape
```

##Calculate Cosine Similarity

```python
similarity = cosine_similarity([resume_embedding],[jd_embedding])
```

```python
print(similarity)
```

```python
match_score = float(similarity[0][0] * 100)
```

```python
print(f"Resume Match Score : {match_score:.2f}%")
```

```python
if match_score >= 85:
      level = "Excellent Match"

elif match_score >= 70:
    level = "Good Match"

elif match_score >= 50:
    level = "Average Match"

else:
    level = "Poor Match"

print(level)
```

```python
print("="*60)

print("Predicted Category :", predicted_category[0])

print("Resume Match Score :", round(match_score,2),"%")

print("Match Quality :", level)

print("="*60)
```

```python
skills_database = [


    # Programming Languages
    "Python", "Java", "C", "C++", "C#", "JavaScript", "TypeScript",
    "R", "Go", "Rust", "PHP", "Kotlin", "Swift",

    # Web Development
    "HTML", "CSS", "Bootstrap", "Tailwind CSS",
    "React", "Angular", "Vue", "Next.js",
    "Node.js", "Express", "Django", "Flask", "FastAPI",
    "Spring", "Hibernate", ".NET", "ASP.NET",

    # Databases
    "SQL", "MySQL", "PostgreSQL", "MongoDB",
    "SQLite", "Oracle", "Redis", "NoSQL",

    # Data Science & ML
    "Machine Learning", "Deep Learning", "NLP",
    "Computer Vision", "CV",
    "Scikit-Learn", "TensorFlow", "Keras",
    "PyTorch", "XGBoost", "LightGBM",

    # Python Libraries
    "NumPy", "Pandas", "Matplotlib",
    "Seaborn", "SciPy", "OpenCV",

    # Big Data
    "Spark", "PySpark", "Hadoop",
    "Hive", "Kafka",

    # Cloud
    "AWS", "Azure", "GCP",
    "EC2", "S3", "Lambda",

    # DevOps
    "Docker", "Kubernetes",
    "Jenkins", "Git", "GitHub",
    "GitLab", "Linux", "Bash",

    # GenAI
    "OpenAI", "LLM", "RAG",
    "LangChain", "LlamaIndex",
    "Transformers", "Hugging Face",
    "FAISS", "ChromaDB",
    "Prompt Engineering",

    # Visualization
    "Power BI", "Tableau",
    "Excel", "Looker",

    # Mobile
    "Android", "Flutter",
    "React Native",

    # Testing
    "JUnit", "Selenium",
    "PyTest", "Unit Testing",

    # Networking & Security
    "Cyber Security",
    "Network Security",
    "Penetration Testing",

    # Software Engineering
    "Agile", "Scrum",
    "Project Management",
    "System Design",
    "REST API",
    "Microservices",

    # Soft Skills
    "Communication",
    "Leadership",
    "Problem Solving",
    "Critical Thinking",
    "Teamwork",
    "Time Management",
    "Presentation Skills",
    "Analytical Thinking"
]

```

```python
import re

def extract_skills(text):

    # Lowercase
        text = text.lower()

    # Replace special characters with space
        text = re.sub(r"[^a-z0-9#+.-]", " ", text)

     # Remove extra spaces
        text = re.sub(r"\s+", " ", text)

        found_skills = set()

        for skill in skills_database:

                skill_lower = skill.lower()


                skill_pattern = re.sub(r"[^a-z0-9#+.-]", " ", skill_lower)

                if skill_pattern in text:

                             found_skills.add(skill)

        return sorted(found_skills)
```

```python
resume_skills = extract_skills(clean_text)

print(resume_skills)
```

```python
jd_skills = extract_skills(clean_jd)

print(jd_skills)
```

```python
matched_skills = list(
      set(resume_skills).intersection(jd_skills)
      )

matched_skills = sorted(matched_skills)

print(matched_skills)
```

```python
missing_skills = list(
      set(jd_skills).difference(resume_skills)
      )

missing_skills = sorted(missing_skills)

print(missing_skills)
```

```python
extra_skills = list(
      set(resume_skills).difference(jd_skills)
      )

extra_skills = sorted(extra_skills)

print(extra_skills)
```

```python
if len(jd_skills)==0:

      skill_match=0

else:

      skill_match=(len(matched_skills)/len(jd_skills))*100

      print(f"Skill Match : {skill_match:.2f}%")
```

```python
print("="*60)

print("Matched Skills")

for skill in matched_skills:

    print("✔",skill)

    print()

print("Missing Skills")

for skill in missing_skills:

        print("❌",skill)

        print()

print("Additional Skills")

for skill in extra_skills:

        print("⭐",skill)

        print()

print(f"Skill Match : {skill_match:.2f}%")

print("="*60)
```
